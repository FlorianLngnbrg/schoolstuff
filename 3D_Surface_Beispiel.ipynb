```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# define function
def f(x, y):
    return np.sin(x) * (1/y)

# generate data
x = np.linspace(0, 20, 100)
y = np.linspace(0, 20, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm')
fig.colorbar(surf)

ax.set_xlabel('t0 [ms]')
ax.set_ylabel('t1 [s]')
ax.set_zlabel('z u. A.')
ax.set_title('3D Fourier-Transformation')

# show plot
plt.show()
```
