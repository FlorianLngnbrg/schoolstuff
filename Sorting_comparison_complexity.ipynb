```python
import numpy as np
import matplotlib.pyplot as plt
import random

# INSERTION SORT
def insertion_sort(arr):
    v, t = 0, 0
    n = len(arr)

    for i in range(1, n):
        j = i
        while j > 0 and arr[j] < arr[j - 1]:
            v += 1
            arr[j], arr[j - 1] = arr[j - 1], arr[j]
            t += 1
            j -= 1
        if j > 0:
            v += 1

    return v, t

# SELECTION SORT
def selection_sort(arr):
    v, t = 0, 0
    n = len(arr)

    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            v += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        t += 1

    return v, t

# BUBBLE SORT
def bubble_sort(arr):
    v, t = 0, 0
    n = len(arr)
    swapped = True

    while swapped:
        swapped = False
        for i in range(n - 1):
            v += 1
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]  # Swap
                t += 1
                swapped = True
        n -= 1

    return v, t


def generate_array(n):
    arr = list(range(1, n + 1))
    random.shuffle(arr)
    return arr

sizes = list(range(5, 1000, 5))
insertion_vergleich = []
selection_vergleich = []
bubble_vergleich = []

for size in sizes:
    arr = generate_array(size)
    insertion_vergleich.append(insertion_sort(arr[:])[0])
    selection_vergleich.append(selection_sort(arr[:])[0])
    bubble_vergleich.append(bubble_sort(arr[:])[0])

# Plot
plt.figure(figsize=(10, 6))
plt.plot(sizes, insertion_vergleich, label='Insertion Sort', marker='o', linestyle='dashed')
plt.plot(sizes, selection_vergleich, label='Selection Sort', marker='s', linestyle='dotted')
plt.plot(sizes, bubble_vergleich, label='Bubble Sort', marker='^', linestyle='solid')
plt.xlabel('Array-Größe')
plt.ylabel('Anzahl der Vergleiche')
plt.title('Vergleich der Sortieralgorithmen')
plt.legend()
plt.grid()
plt.show()
```
