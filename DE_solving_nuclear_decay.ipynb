```python
import matplotlib.pyplot as plt

# Rahmenbedingungen
NM_s = 200
NT_s = 0

lM = 0.02
lT = 0.03

def n_tm(t, dt):
  return NM[t] - NM[t] * lM * dt

def n_tn(t, dt):
  return  NT[t] + dt * (NM[t] * lM - NT[t] * lT)

n = 100

NM = [NM_s]
NT = [NT_s]

x = [i for i in range(101)]

# Simulation
for i in range(n):
  NM.append(n_tm(i, 1))
  NT.append(n_tn(i, 1))

plt.scatter(x, NM)
plt.scatter(x, NT)

plt.show()
```
